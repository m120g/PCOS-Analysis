**Heatmap of the Dataset**

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Convert non-numeric values to NaN for all columns
df = df.apply(pd.to_numeric, errors='coerce')

# Drop rows with any NaN values
df = df.dropna()

# Calculate correlation matrix
corrmat = df.corr()

k = 12  # number of variables with positive correlation for heatmap
l = 3   # number of variables with negative correlation for heatmap

cols_p = corrmat.nlargest(k, 'PCOS (Y/N)')['PCOS (Y/N)'].index 
cols_n = corrmat.nsmallest(l, 'PCOS (Y/N)')['PCOS (Y/N)'].index
cols = cols_p.append(cols_n)

cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
plt.figure(figsize=(12, 12))
cmap = ListedColormap(pink_palette)

hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10},
                 yticklabels=cols.values, xticklabels=cols.values,cmap=cmap)
plt.show()
```

*Displays the correlation between all the variables present in the dataset. The correlation between the left and righ tfollicle is the highest amongst all.*
